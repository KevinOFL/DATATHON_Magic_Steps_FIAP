## Funções reutilizaveis

* Função de plotagem de gráfico de barras com diferencial entre notas dos evadidos que estão acima de mediana geral no indice e abaixo com visualização da quantidade.

```python
# Plota um gráfico de barras para a coluna desejada de um DataFrame
def plot_grafico_barras_indices_evadidos(df, indice_coluna, mediana_geral, titulo_grafico, xlabel, ylabel, figsize=(9, 12)):
    # Lógica para definir as cores das barras com base na média
    ordem_notas = df[indice_coluna].value_counts().index

    cores = ['#4CAF50' if nota >= mediana_geral else '#F44336' for nota in ordem_notas]

    plt.figure(figsize=figsize)

    ax = sns.countplot(
        data=df,
        y=indice_coluna,
        order=ordem_notas,
        palette=cores
    )

    for container in ax.containers:
        labels = [f"{int(valor)}" if valor > 0 else "" for valor in container.datavalues]
        ax.bar_label(container, labels=labels, label_type='edge', padding=3, fontsize=10)
        
    max_count = df[indice_coluna].value_counts().max()
    plt.xlim(0, max_count * 1.1)
    plt.title(titulo_grafico)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)

    plt.xticks(rotation=0)
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```

* Função que retorna a porcentagem de alunos que estão abaixo e acima da mediana geral

```python
def porcentagem_acima_e_abaixo_da_mediana(df, indice_coluna, mediana_geral):
    cont = len(df)
    
    if cont == 0:
        return 0.0, 0.0
    
    acima_da_mediana = df[df[indice_coluna] >= mediana_geral]
    abaixo_da_mediana = df[df[indice_coluna] < mediana_geral]

    porcentagem_acima = (len(acima_da_mediana) / cont) * 100
    porcentagem_abaixo = (len(abaixo_da_mediana) / cont) * 100

    return porcentagem_acima, porcentagem_abaixo
```

* Função que gera a mediana de uma coluna especifica de um dataframe

```python
def mediana_indice(df, indice_coluna):
    return df[indice_coluna].median()
```

* Função que efetua o print de estátisticas de todos os evadidos

```python
def print_estatisticas_evadidos_geral(df_geral, df_alvo, indice_coluna, ano_referencia):
    mediana_geral = df_geral[indice_coluna].median()
    mediana_alvo = df_alvo[indice_coluna].median()
    porcentagem_acima, porcentagem_abaixo = porcentagem_acima_e_abaixo_da_mediana(df_alvo, indice_coluna, mediana_geral)
    
    print(f"Mediana geral de {indice_coluna} para {ano_referencia} dos RAs que permaneceram: {mediana_geral}")
    print(f"Mediana de {indice_coluna} para RAs presentes em {ano_referencia} mas não em {ano_referencia+1}: {mediana_alvo}")
    print(f"Porcentagem de RAs acima da mediana: {porcentagem_acima:.1f}%")
    print(f"Porcentagem de RAs abaixo da mediana: {porcentagem_abaixo:.1f}%")
```

* Função que efetua o print de estátisticas de todos os evadidos por gênero

```python
def print_estatisticas_evadidos_por_genero(df_geral, df_alvo, indice_coluna, ano_referencia):
    mediana_geral = df_geral[indice_coluna].median()
    if ano_referencia == 2022:
        df_fem = df_alvo[df_alvo['Gênero'] == 'Menina'].copy()
        df_mac = df_alvo[df_alvo['Gênero'] == 'Menino'].copy()
        porcentagem_abaixo_fem, porcentagem_acima_fem = porcentagem_acima_e_abaixo_da_mediana(df_fem, indice_coluna, mediana_geral)
        porcentagem_abaixo_mac, porcentagem_acima_mac = porcentagem_acima_e_abaixo_da_mediana(df_mac, indice_coluna, mediana_geral)
        print(f"Quantidade total de RAs feminino que evadiram: {len(df_fem)}")
        print(f"Quantidade total de RAs masculino que evadiram: {len(df_mac)}")
        print(f"Porcentagem de RAs que evadiram do gênero Feminino acima da mediana: {porcentagem_acima_fem:.1f}%")
        print(f"Porcentagem de RAs que evadiram do gênero Feminino abaixo da mediana: {porcentagem_abaixo_fem:.1f}%")
        print(f"Porcentagem de RAs que evadiram do gênero Masculino acima da mediana: {porcentagem_acima_mac:.1f}%")
        print(f"Porcentagem de RAs que evadiram do gênero Masculino abaixo da mediana: {porcentagem_abaixo_mac:.1f}%")
    elif ano_referencia == 2023:
        df_fem = df_alvo[df_alvo['Gênero'] == 'Feminino'].copy()
        df_mac = df_alvo[df_alvo['Gênero'] == 'Masculino'].copy()
        porcentagem_abaixo_fem, porcentagem_acima_fem = porcentagem_acima_e_abaixo_da_mediana(df_fem, indice_coluna, mediana_geral)
        porcentagem_abaixo_mac, porcentagem_acima_mac = porcentagem_acima_e_abaixo_da_mediana(df_mac, indice_coluna, mediana_geral)
        print(f"Quantidade total de RAs feminino que evadiram: {len(df_fem)}")
        print(f"Quantidade total de RAs masculino que evadiram: {len(df_mac)}")
        print(f"Porcentagem de RAs que evadiram do gênero Feminino acima da mediana: {porcentagem_acima_fem:.1f}%")
        print(f"Porcentagem de RAs que evadiram do gênero Feminino abaixo da mediana: {porcentagem_abaixo_fem:.1f}%")
        print(f"Porcentagem de RAs que evadiram do gênero Masculino acima da mediana: {porcentagem_acima_mac:.1f}%")
        print(f"Porcentagem de RAs que evadiram do gênero Masculino abaixo da mediana: {porcentagem_abaixo_mac:.1f}%")
```
